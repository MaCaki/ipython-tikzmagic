```python
%%tikz -l arrows,matrix
\matrix (m) [matrix of math nodes, row sep=3em, column sep=4em] {
A & B \\
C & D \\
};
\path[-stealth, line width=.4mm]
(m-1-1) edge node [left ] {$ac$} (m-2-1)
(m-1-1) edge node [above] {$ab$} (m-1-2)
(m-1-2) edge node [right] {$bd$} (m-2-2)
(m-2-1) edge node [below] {$cd$} (m-2-2);
```

Using the tkz-berge package you can make nice looking graphs, with a caption. 

```python
%%tikz -s 800,600 -sc 1.2 -l arrows,petri,topaths -p tkz-berge -c "A directed graph."

\Vertex[x=0,y=4]{A}
\Vertex[x=2,y=2]{B}
\Vertex[x=0,y=0]{C}
\tikzstyle{LabelStyle}=[fill=white,sloped]
\tikzstyle{EdgeStyle}=[post,bend right]
\Edge[style={blue}, label=a,labelstyle={left}](A)(B)
\Edge[style={blue}, label=b,labelstyle={left}](B)(C)

\tikzset{EdgeStyle/.style={->}}
\tikzstyle{EdgeStyle}=[post, bend right]
\Edge[style={red}](A)(C)
\Loop[style={red}, dir=SO, label=l](B)(B)
\tikzstyle{EdgeStyle}=[post, bend right]
\Edge[style={red}](C)(B)
```
